```python
def quanQual(dataset):
    quan=[]
    qual=[]
    for columnName in dataset.columns:
        #print(columnName)
        if(dataset[columnName].dtype=='O'):
            #print("qual")
            qual.append(columnName)
        else:
            #print("quan")
            quan.append(columnName)
    return quan,qual
```

```python
quan,qual=quanQual(bank_fraud_df)
```

```python
quan
```

```python
qual
```

```python
descriptive_all_data=pd.DataFrame(index=["Mean","Median","Mode","Min","25%","50%","75%","100%","IQR",
                               "1.5IQR","LesserRange","GreaterRange"],columns=quan)
for columnName in quan:
    descriptive_all_data[columnName]["Mean"]=bank_fraud_df[columnName].mean()
    descriptive_all_data[columnName]["Median"]=bank_fraud_df[columnName].median()
    descriptive_all_data[columnName]["Mode"]=bank_fraud_df[columnName].mode()[0]
    descriptive_all_data[columnName]["Min"]=bank_fraud_df[columnName].min()
    descriptive_all_data[columnName]["25%"]=bank_fraud_df.describe()[columnName]["25%"]
    descriptive_all_data[columnName]["50%"]=bank_fraud_df.describe()[columnName]["50%"]
    descriptive_all_data[columnName]["75%"]=bank_fraud_df.describe()[columnName]["75%"]
    descriptive_all_data[columnName]["100%"]=bank_fraud_df.describe()[columnName]["max"]
    descriptive_all_data[columnName]["IQR"]=descriptive_all_data[columnName]["75%"]-descriptive_all_data[columnName]["25%"]
    descriptive_all_data[columnName]["1.5IQR"]=1.5*descriptive_all_data[columnName]["IQR"]
    descriptive_all_data[columnName]["LesserRange"]=descriptive_all_data[columnName]["25%"]- descriptive_all_data[columnName]["1.5IQR"]
    descriptive_all_data[columnName]["GreaterRange"]=descriptive_all_data[columnName]["75%"]+ descriptive_all_data[columnName]["1.5IQR"]
```

```python
#Load  the column value to the independent variable (input data)
independent=bank_fraud_df[['Transaction_Amount', 'Account_Balance']]
```

```python
#Load  the column value to the dependent variable (output data)
dependent=bank_fraud_df[['Is_Fraud']]
```

```python
#Scikit-learn is an open-source machine learning library.It is a powerful tool for classification, regression, clustering, and dimensionality reduction.
#Split split the data into Training Data set  and Test Data Set
#X - Input, y -Output
#Load  the  function  from the sklearn library (SKLEARN provide the library for Machine Language)
#test_size is set to 0.50 %
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
# Model Creation using Lienar Regression  (Numerical Output)
# Key concept in model creation is comeup with Good value for  Weight and  bais.
# Load LinearyRegression class to the variable regressor 
# The following  first two lines  are procedure
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
# Executing the Fit function  using the training  data set (Model created)- Weight and  Bias is calculated
regressor.fit(X_train,y_train)
```

```python
# Find the weight calculated using the  training  data
# m=slope=dy/dx=weight(Which says about the constant distance between two data points)
#.coef  returns the value for weight
weight=regressor.coef_
weight
```

```python
# Find the bais  calculated using the  training data
# c=bias=intercpet=stating of the strait line=initial value= minimum value
# intercept  returns  the bias
bais=regressor.intercept_
bais
```

```python
# Find the prediction  for the input value X_test
y_pred=regressor.predict(X_test)
```

```python
# Evaluvate the test output (actual) against prediction  to validate the model
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
```

```python
#r_score  value is  closer to  number 1  then it indicates the model is  good  model
r_score
```
