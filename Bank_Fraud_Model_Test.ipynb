```python
import pickle
import pandas as pd
```

# Input Columns
'Age', 'Transaction_Amount', 'Account_Balance', 'Transaction_Type', 'Merchant_Category', 'Device_Type', 'State', 'City'

#  Load & Test the Model

```python
# Load saved model
rf_model = pickle.load(open('Bank_Fraud_model_RF.sav', 'rb'))
```

```python
# New fRAUD transaction test data
new_transaction = pd.DataFrame([{
    'Age':31,
    'State': 'Assam',
    'City': 'Delhi',
    'Transaction_Amount':91610,
    'Transaction_Type': 'Credit',
    'Merchant_Category': 'Health',
    'Account_Balance': 83254,
    'Device_Type': 'POS'   
  }])
# Prediction
prediction = rf_model.predict(new_transaction)
print("Fraud Prediction:", prediction)
# Probability
probability = rf_model.predict_proba(new_transaction)
print("Fraud Probability:", probability[0][1])
```

```python
# Threshold decision
if probability[0][1] >= 0.3:
    print("Fraud Transaction")
else:
    print("Safe Transaction")
```
