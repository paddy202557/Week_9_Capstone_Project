```python
import pickle
import pandas as pd
import numpy as np
```

# Input Columns
'Age', 'Transaction_Amount', 'Account_Balance', 'Transaction_Type', 'Merchant_Category', 'Device_Type', 'State', 'City'

#  Load & Test the Model

```python
# Load saved mode
rf_model  = pickle.load(open('Bank_Fraud_finalized_model_RF.sav', 'rb'))
knn_model = pickle.load(open('Bank_Fraud_finalized_model_KNN.sav', 'rb'))
bnb_model = pickle.load(open('Bank_Fraud_finalized_model_BNB.sav', 'rb'))
```

# Fradulent transaction test data

```python
# Fradulent transaction test data
new_transaction = pd.DataFrame([{
    'Age':56,
    'State': 'Dadra and Nagar Haveli and Daman and Diu',
    'City': 'Silvassa',
    'Transaction_Amount':85269.30,
    'Transaction_Type': 'Debit',
    'Merchant_Category': 'Restaurant',
    'Account_Balance':74322.85,
    'Device_Type': 'Desktop'   
  }])
```

# Random Forest model prediction

```python
# Prediction
print("\033[1;31mRandom Forest model prediction\033[0m")
prediction = rf_model.predict(new_transaction)
print("Fraud Prediction:", prediction)
# Probability
probability = rf_model.predict_proba(new_transaction)
print("Fraud Probability:", probability[0][1])

# Threshold decision
if probability[0][1] >= 0.3:
    print("Transaction is Fraud")
else:
    print("Transaction is Safe")
```

# K-Nearest Neighbour model prediction

```python
# Prediction
print("\033[1;31mK-Nearest Neighbour model prediction\033[0m")
prediction = knn_model.predict(new_transaction)
print("Fraud Prediction:", prediction)
# Probability
probability = knn_model.predict_proba(new_transaction)
print("Fraud Probability        :", probability[0][1])
probability = np.round(knn_model.predict_proba(new_transaction),1)
print("Fraud Probability rounded:", probability[0][1])
# Threshold decision
if probability[0][1] >= 0.3:
    print("Transaction is Fraud")
else:
    print("Transaction is Safe")

```

# Bernoulli Naive Bias model predciton

```python
# Prediction
print("\033[1;31m K-Nearest Neighbour model prediction\033[0m")
prediction = bnb_model.predict(new_transaction)
print("Fraud Prediction:", prediction)
# Probability
probability = bnb_model.predict_proba(new_transaction)
print("Fraud Probability        :", probability[0][1])
probability = np.round(bnb_model.predict_proba(new_transaction),1)
print("Fraud Probability rounded:", probability[0][1])
# Threshold decision
if probability[0][1] >= 0.3:
    print("Transaction is Fraud")
else:
    print("Transaction is Safe")
```
